```python
import pandas as pd
```

```python
df_etape = pd.read_csv("data/DetailEtape.csv")
df_etape.columns
```

On crée une liste de colonnes représentants les indicateurs écologiques  
Puis on calcule la moyenne par groupe d'aliment

```python
liste_indicateurs = df_etape.columns[10:16].values
df_etape.groupby("Groupe d\'aliment")[liste_indicateurs].mean()
```

On calcule la moyenne par groupe et sous-groupe d'aliment

```python
df_etape.groupby(["Groupe d\'aliment","Sous-groupe d'aliment"])[liste_indicateurs].mean()
```

Focus sur viandes, oeufs et poissons

```python
df_etape[df_etape["Groupe d\'aliment"] == "viandes, œufs, poissons"].groupby(["Groupe d\'aliment","Sous-groupe d'aliment"])[liste_indicateurs].mean()
```

Focus sur viandes crues

```python
df_etape[df_etape["Sous-groupe d'aliment"] == "viandes crues"].groupby(["Groupe d\'aliment","Sous-groupe d'aliment"])[liste_indicateurs].mean()
```

On regroupe par type d'animal

```python
especes = ["veau", "boeuf", "poulet", "agneau", "canard", "porc", "oie", "mouton", "lapin", "génisse", "dinde"]
espece = especes[0]
produit_espece = df_etape['Nom du Produit en Français'].str.contains(espece).sum()
produit_espece
```

```python
for espece in especes:
    df_etape.loc[df_etape['Nom du Produit en Français'].str.contains(espece),'Espece'] = espece.capitalize()
```

```python
df_etape.groupby('Espece')[liste_indicateurs].mean()
```
