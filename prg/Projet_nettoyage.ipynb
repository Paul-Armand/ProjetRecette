## Audit des données

```python
import pandas as pd
```

```python
df_synthese = pd.read_csv("data/Synthese.csv")
df_synthese.head(3)
```

```python
df_synthese.info
```

Je regarde s'il y a des doublons entre les 2 premières colonnes

```python
nombre_doublons = 0
for index, ligne in df_synthese.iterrows():
    if int(ligne[0]) != int(ligne[1]):
        #print(f"La colonne 1 vaut {ligne[0]} et la colonne 2 vaut {ligne[1]}")
        print(f"Il y a une différence à l'index {index} : {ligne[0]} vs {ligne[1]}")
        nombre_doublons += 1
print(f"Il y a {nombre_doublons} doublons")
```

```python
# On décide de supprimer la colonne AGB
# df_synthese.drop(columns='Code\nAGB', inplace=True)
```

```python
df_synthese.to_csv("data/SyntheseSansAGB.csv", index=False)
```

```python
df_detailetape = pd.read_csv("data/DetailEtape.csv")
df_detailetape.info
```

```python
nombre_doublons = 0
for index,ligne in df_detailetape.iterrows():
    if int(ligne[0]) != int(ligne[1]):
        nombre_doublons += 1
        print(f"Il y a une différence à l'index {index} : {ligne[0]} vs {ligne[1]}")
nombre_doublons
```

```python
# On constate la même chose que pour la synthèse. On décide de supprimer la colonne AGB
df_detailetape.drop(columns='Code AGB', inplace=True)
```

```python
df_detailetape.to_csv("data/DetailEtapeSansAGB.csv", index=False)
```
